# Análisis Exploratorio de Datos (EDA)
## Conjunto de datos: Anuncios de venta de coches usados

```python
import pandas as pd
import plotly.express as px
```

```python
# Leer el conjunto de datos
car_data = pd.read_csv('../vehicles_us.csv')
car_data.head()
```

```python
# Información general del dataset
car_data.info()
```

```python
# Estadísticas descriptivas
car_data.describe()
```

```python
# Histograma del kilometraje
fig = px.histogram(car_data, x='odometer')
fig.update_layout(title_text='Distribución del Odómetro')
fig.show()
```

```python
# Gráfico de dispersión: Precio vs Kilometraje
fig = px.scatter(car_data, x='odometer', y='price')
fig.update_layout(title_text='Relación entre Odómetro y Precio')
fig.show()
```

```python
# Histograma de precios
fig = px.histogram(car_data, x='price')
fig.update_layout(title_text='Distribución de Precios')
fig.show()
```

```python
# Histograma por tipo de vehículo
fig = px.histogram(car_data, x='type')
fig.update_layout(title_text='Distribución por Tipo de Vehículo')
fig.show()
```
